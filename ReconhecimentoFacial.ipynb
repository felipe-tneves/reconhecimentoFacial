# **Reconhecimento facial**

### O código abaixo foi baseado apartir dessa publicação: http://iaacademy.com.br/2020/05/07/reconhecimento-de-faces-com-python-e-opencv/ 

```python
#Importando bibliotecas
import cv2
from google.colab import files
from google.colab.patches import cv2_imshow 
```

```python
#Importando a imagem
uploaded = files.upload()
```

```python
#Lendo a imagem
img = cv2.imread('formados.jpg')
```

```python
#Covertendo a imagem para escalas de cinzas
imgCinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#O arquivo pode ser encontrado nesse link: https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_alt.xml
#Carregar o arquivo XML com os dados obtidos por meio de um treinamento prévio para a identificação de faces
classificador = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
#Lista contendo a localização de cada face identificada.
#Cada localização é uma lista contendo as posições X e Y de cada retângulo com uma face identificada, bem como a largura e altura de cada um.
deteccoes = classificador.detectMultiScale(imgCinza)
```

```python
#Varreremos todos os itens da variável “deteccoes” e efetivamente desenharemos os retângulos que delimitam cada face na versão colorida de nossa imagem.
for (x, y, l, a) in deteccoes:
    cv2.rectangle(img, (x, y), (x + l, y + a), (0, 255, 0), 2)
```

```python
#Apresenta a imagem com a identificação dos rostos
cv2_imshow(img)
```
